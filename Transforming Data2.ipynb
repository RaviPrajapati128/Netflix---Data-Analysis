#### 1) A Box-office hit film means a film which has grossed $20,00,000. Find the number of Box-Office Hit Films.

#### 2) A critically acclaimed movie means a film that has IMDB rating of at least 8. Find the number of such films.

```python
critically_acclaimed = data[data["IMDb Score"]>=8]
print("critically_acclaimed:",critically_acclaimed["Title"].count())
```

#### 3) Find all critically acclaimed movies which were Box-Office Hits.

```python
m = data[(data["IMDb Score"]>=8) & (data["Gross"]>=2000000)].Title.count()
print("Hit & Critically Claimed Movies:",m)
```

#### 4) Find all critically acclaimed movies which has received at least 5 awards. Order them according to the number of awards.

```python
movie = data[(data["IMDb Score"]>=8) & (data["Awards Nominated For"]>=5)]
print(movie[["Title","Awards Nominated For"]].sort_values(by = "Awards Nominated For"))
```

#### 7) A popular film means it is released at Netflix platforms within 1 year. Find all Films which were released in Netflix within 2 years of actual release.

```python
data['Release Date'] = pd.to_datetime(data['Release Date'], format='%d/%m/%Y', errors='coerce')
data['Netflix Release Date'] = pd.to_datetime(data['Netflix Release Date'], format='%d/%m/%Y', errors='coerce')

popular_films = data[(data['Series or Movie'] == 'Movie') & ((data['Netflix Release Date'] - data['Release Date']).dt.days <= 365)]
popular_films.head()
```

#### 8) Find all popular films which are available in Canada.

```python
popular_films_canada = popular_films[popular_films['Country Availability'].str.contains('Canada', na=False)]
popular_films_canada
```
